```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.linspace(0,10,11)
```

```python
m
```

```python
c = 3 * 10**8 
```

```python
c
```

```python
E = m * c**2
```

```python
E
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,0.5,0.5])
axes.plot(m, E,color="red",lw=3)

axes.set_title("E=mc^2")
axes.set_xlabel("Mass in grams",fontsize=8)
axes.set_ylabel("Energy in Joules",fontsize=8)
axes.set_xlim(0,10)



plt.show()
```
