```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y =2*x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xlim(0,6) #can also give limits
plt.ylim(0,15)
plt.title('Title');
plt.savefig('myfirstplot.png')
```

# FIGURES AND AXES

```python
a = np.linspace(0,10,11)
a
```

```python
b = a**4
b
```

```python
fig = plt.figure(figsize=(4,4),dpi=100)

#big axes
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)


#small axes
axes2 = fig.add_axes([0.2,0.5,0.2,0.2])
axes2.plot(x,y)


fig.savefig('myfig.png',bbox_inches='tight')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_title('Power of 4');

axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(a,b)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_title('Zoomed In')
```
