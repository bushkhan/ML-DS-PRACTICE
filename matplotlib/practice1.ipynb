```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y = x*2
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(0,6)
plt.ylim(0,15)

plt.title("Title")
plt.savefig('practiceplot.png')
```

```python
fig = plt.figure(dpi=100)

a = np.linspace(0,10,11)
b = a**2

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes2 = fig.add_axes([0.1,0.5,0.25,0.25])
axes2.plot(a,b)

fig.savefig("practice_fig.png",bbox_inches = "tight")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
a = np.arange(0,10)
b = a ** 4

x = np.linspace(0,10,11)
y = 2 * x
print(a)
axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
a = np.arange(0,10)
b = a ** 4

x = np.linspace(0,10,11)
y = 2 * x
print(a)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)

fig.suptitle("FIG LEVELE TITLE")
# fig.subplots_adjust(wspace=0.6,hspace=0.5)
plt.tight_layout()
```
