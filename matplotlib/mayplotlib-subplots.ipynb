```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a ** 2
```

```python
x = np.arange(1,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2) #tuple-unpacking #can give rows and columns here
```

```python
type(axes)
```

```python
axes.shape
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2) #2D
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(a,b)
axes[1][1].plot(a,b)

axes[0][0].set_title("Fig 1")
axes[1][0].set_xlabel("x axis")
fig.suptitle("Figure Level",fontsize=16,fontweight=13)

fig.subplots_adjust(wspace=1,hspace=0.6) #width space and height space

# fig.set_title()figure level options also available
fig.savefig("Subplots example",bbox_inches='tight')
```

```python
fig,axes = plt.subplots(nrows=3) #3rows

for ax in axes:
    ax.plot(x,y)
    
fig.tight_layout()
```

# STYLING:

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,x,label='X vs X',color='red',lw=2,ls='--') #lw = line width ls=line style
axes1.plot(x,x*2,label='X vs X^2',color='#e6c017',marker='o',markersize=4)
axes1.legend(loc=0)
plt.xlim(1,10)
plt.ylim(1,15)
```
