```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df,x='parental level of education',y='math score',hue='test preparation course')
plt.legend(bbox_to_anchor = (1.05,1))
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,y='parental level of education',x='math score',hue='test preparation course')
plt.legend(bbox_to_anchor = (1.05,1),title='test preparation course')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.violinplot(data=df,y='parental level of education',x='math score',hue='test preparation course',
               palette='Set2')
plt.legend(bbox_to_anchor = (1.05,1),title='test preparation course')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.violinplot(data=df,y='parental level of education',x='math score',split=True,inner='quartile')
plt.show()
```

```python
sns.swarmplot(data=df,x='math score',size=2)
plt.show()
```

```python
sns.swarmplot(data=df,x='math score',y='gender',size=2,hue='test preparation course',dodge=True)
plt.legend(bbox_to_anchor = (1.05,1),title='test preparation course')

plt.show()
```
