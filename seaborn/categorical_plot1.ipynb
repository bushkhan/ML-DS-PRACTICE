```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df['division'].value_counts() #we can only have counts when there are categories thats useful just like 
#division column and leel of education column where data is divided into categories
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='division')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='level of education')
#plt.ylim(-10,500)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='division',hue='level of education',palette='Set2')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='level of education', y='salary',data=df,estimator=np.mean,ci='sd') 
#mean of salary shown below with standard daviation
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='level of education', y='salary',data=df,estimator=np.mean,ci='sd',hue='division') 
#mean of salary shown below with standard daviation
plt.legend(bbox_to_anchor = (1.05,1))
plt.show()

```
