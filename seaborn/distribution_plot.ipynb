```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.rugplot(x='salary',data=df, height=0.4)
plt.show()
```

```python
sns.kdeplot(x='salary',data=df)
plt.show()
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=20,kde=True,rug=True) #all in one kde also rug also in displot
#edgecolor
#linewidth
#color
#ls
plt.show()
```

```python
sns.histplot(x='salary',data=df)
```
