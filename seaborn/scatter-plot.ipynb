```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df
```

```python
sns.scatterplot(x='salary', y='sales',data=df)
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary', y='sales',data=df)
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary', y='sales',data=df, hue='division',palette='Dark2',size='salary',alpha=1)
plt.savefig("scatter_plot.png")
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary', y='sales',data=df, style='level of education', hue='level of education')
plt.show()
```
